```python
import math

class Point:
    '''
    Represents a point in 2D space

    attributes: x, y
    '''

    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y

    def __str__(self):
        return f'({self.x},{self.y})'


class Circle:


    def __init__(self, input_radius = 0, input_center = Point()):
        self.radius = input_radius
        self.center = input_center


    def __str__(self):
        return f'radius: {self.radius} , center:{self.center}'
    
    def get_area(self):
        area = math.pi * ((self.radius)** 2)
        return area
    

class Rectangle:


    def __init__(self, input_point = Point(), input_width = 0,  input_height = 0):
        self.corner = input_point
        self.width = input_width
        self.height = input_height


    def __str__(self):
        return f'Corner: {self.corner}, Width: {self.width}, Height: {self.height}'


    def get_area(self):

        area = self.width * self.height
        return area

    def get_center_Point(self):

        center_x = self.corner.x + self.width / 2
        center_y = self.corner.y + self.height / 2
        return Point(center_x, center_y)
    
    def to_circle(self):

        area = self.get_area()
        radius = math.sqrt(area / math.pi)
        center_point = self.get_center_Point()
        return f'radius:{radius}, center:{center_point}'
    
ExRectangle = Rectangle(Point(), 6, 4)
RectToCricle = ExRectangle.to_circle()
print(RectToCricle)
    
```

```python
import math

class Point:
    '''
    Represents a point in 2D space

    attributes: x, y
    '''

    def __init__(self, input_x = 0, input_y = 0):
        self.x = input_x
        self.y = input_y

    def __str__(self):
        return f'({self.x},{self.y})'

class Rectangle:


    def __init__(self, input_point = Point(), input_width = 0,  input_height = 0):
        self.corner = input_point
        self.width = input_width
        self.height = input_height


    def __str__(self):
        return f'Corner: {self.corner}, Width: {self.width}, Height: {self.height}'


    def get_area(self):

        area = self.width * self.height
        return area

    def get_center_Point(self):

        center_x = self.corner.x + self.width / 2
        center_y = self.corner.y + self.height / 2
        return Point(center_x, center_y)
    
    def to_circle(self):

        area = self.get_area()
        radius = math.sqrt(area / math.pi)
        center_point = self.get_center_Point()
        return f'radius:{radius}, center:{center_point}'


class Circle:


    def __init__(self, input_radius = 0, input_center = Point()):
        self.radius = input_radius
        self.center = input_center


    def __str__(self):
        return f'radius: {self.radius} , center:{self.center}'
    
    def get_area(self):
        area = math.pi * ((self.radius)** 2)
        return area

    def to_square(self):
        area = self.get_area()
        sides = math.sqrt(area)
        x = self.center.x - (sides / 2)
        y = self.center.y - (sides / 2)
        corner = Point(x,y)
        return f'Corner: {corner} Width: {sides} Height: {sides}'

    
#test case 2
ExCircle = Circle(1, Point())

square = ExCircle.to_square()
print(square)
    

```
