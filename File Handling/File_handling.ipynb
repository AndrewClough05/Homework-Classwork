```python
# Question 1
def histogram(string_list):
    string_dict = dict()

    for i in string_list:
        if i in string_dict:
            string_dict[i] += 1
        else:
            string_dict[i] = 1
            
    return string_dict

def anograms(input_string):
     with open('words.txt', 'r') as read_file: 
         with open('output.txt', 'w') as write_file: 
             for line in read_file:
                 
                line = line.strip()
         
                if histogram(input_string) == histogram(line):
                    write_file.write(line + '\n')


                 
                 

                 
    
```

```python
anograms("pots")
```
